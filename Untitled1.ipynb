```python
#@markdown To better demonstrate the Pose Landmarker API, we have created a set of visualization tools that will be used in this colab. These will draw the landmarks on a detect person, as well as the expected connections between those markers.

from mediapipe import solutions
from mediapipe.framework.formats import landmark_pb2
import numpy as np


def draw_landmarks_on_image(rgb_image, detection_result):
  pose_landmarks_list = detection_result.pose_landmarks
  annotated_image = np.copy(rgb_image)

  # Loop through the detected poses to visualize.
  for idx in range(len(pose_landmarks_list)):
    pose_landmarks = pose_landmarks_list[idx]

    # Draw the pose landmarks.
    pose_landmarks_proto = landmark_pb2.NormalizedLandmarkList()
    pose_landmarks_proto.landmark.extend([
      landmark_pb2.NormalizedLandmark(x=landmark.x, y=landmark.y, z=landmark.z) for landmark in pose_landmarks
    ])

    import copy
    style = copy.deepcopy(solutions.drawing_styles.get_default_pose_landmarks_style())
    for k in style:
      style[k].circle_radius = 10
      style[k].thickness = 10
    solutions.drawing_utils.draw_landmarks(
      annotated_image,
      pose_landmarks_proto,
      solutions.pose.POSE_CONNECTIONS,
      style)
  return annotated_image
```

```python
import math
from typing import Tuple
import numpy as np

def calculate_angle_3d(mark1, mark2, mark3):

    A = np.array([mark1.x, mark1.y, mark1.z])
    B = np.array([mark2.x, mark2.y, mark2.z])
    C = np.array([mark3.x, mark3.y, mark3.z])


    # build direction vectors BA and BC
    v1 = A - B
    v2 = C - B

    # compute unit‐cross for the rotation axis
    axis = np.cross(v1, v2)
    axis /= np.linalg.norm(axis)
    print(f"Axis of rotation: {axis}")

    # dot / (|v1|*|v2|) → clamp → arccos → interior angle
    cosθ = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
    cosθ = np.clip(cosθ, -1.0, 1.0)
    θ = np.degrees(np.arccos(cosθ))

    # if you really want the *supplementary* (180–θ), do: return 180-θ
    return (180-θ)
```

```python
def calculate_angle_2d(mark1, mark2, mark3):
    # build NumPy vectors from landmark objects
    A = np.array([mark1.x, mark1.y])
    B = np.array([mark2.x, mark2.y])
    C = np.array([mark3.x, mark3.y])

    # direction vectors BA and BC
    v1 = A - B
    v2 = C - B

    # compute cosine of the angle, clamp to [-1,1]
    cosθ = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
    cosθ = np.clip(cosθ, -1.0, 1.0)

    # θ in degrees
    θ = np.degrees(np.arccos(cosθ))

    # return supplementary angle (180°−θ)
    return 180.0 - θ
```

```python
import numpy as np

def calculate_knee_angle(
    hip, knee, ankle,
    angle_type: str = "interior",
    reference_axis: np.ndarray = None
) -> float:
    """
    Calculate the knee angle at `knee`, formed by segments hip→knee→ankle.

    Parameters
    ----------
    hip, knee, ankle : objects with .x, .y, .z attributes
        3D coordinates of the joint centers.

    angle_type : {"interior", "exterior", "signed"}
        - "interior": the smaller angle between thigh and shank (0°–180°).
        - "exterior": the reflex angle (180°–360°), i.e. 360° – interior.
        - "signed": interior angle, but positive vs. negative by projecting
          the cross‐product onto `reference_axis`.

    reference_axis : array‐like, shape (3,), optional
        Required if angle_type == "signed".  Defines your mediolateral axis
        (e.g. pelvis left→right) so that:
          • +θ → flexion,
          • –θ → hyperextension (or vice versa, depending on how you set it).

    Returns
    -------
    float
        The requested knee angle, in degrees.
    """

    # --- 1) build vectors ---
    H = np.array([hip.x, hip.y, hip.z], dtype=float)
    K = np.array([knee.x, knee.y, knee.z], dtype=float)
    A = np.array([ankle.x, ankle.y, ankle.z], dtype=float)

    v_thigh = H - K
    v_shank = A - K

    # --- 2) interior angle via dot/cos ---
    dot = np.dot(v_thigh, v_shank)
    denom = np.linalg.norm(v_thigh) * np.linalg.norm(v_shank)
    cosθ = np.clip(dot / denom, -1.0, 1.0)
    θ_int = np.degrees(np.arccos(cosθ))  # in [0, 180]

    # --- 3) return as requested ---
    if angle_type == "interior":
        return 180.0 - θ_int

    if angle_type == "exterior":
        return 360.0 - θ_int

    if angle_type == "signed":
        if reference_axis is None:
            raise ValueError("reference_axis is required for signed angles")

        # axis from thigh→shank
        axis = np.cross(v_thigh, v_shank)
        norm_axis = np.linalg.norm(axis)
        if norm_axis < 1e-8:
            # nearly colinear → no meaningful sign
            return 0.0
        axis_unit = axis / norm_axis

        # project onto your anatomical axis
        ref = np.asarray(reference_axis, dtype=float)
        ref_unit = ref / np.linalg.norm(ref)
        sign = np.sign(np.dot(axis_unit, ref_unit))
        return sign * θ_int

    raise ValueError("angle_type must be 'interior', 'exterior' or 'signed'")
```
